```python
import pandas as pd
import boto3
from io import StringIO

s3 = boto3.client('s3')
bucket_name = 'cmse492-chaud114-nyc311-704788404995-us-east-1-an' # replace with your bucket name (make sure you have the account regional suffix)
file_name = 'modeling/resolution_time_modeling.csv' # make sure to include the path if it's in a folder, e.g. 'modeling/modeling_data.csv'

obj = s3.get_object(Bucket=bucket_name, Key=file_name)
data = obj['Body'].read().decode('utf-8')
df = pd.read_csv(StringIO(data))
```
